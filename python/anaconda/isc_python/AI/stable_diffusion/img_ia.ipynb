```python
import tensorflow as tf
tf.keras.utils.set_random_seed(42)  # Fixez une graine pour éviter les incohérences
tf.config.experimental.enable_tensor_float_32_execution(False)  # Désactive certaines optimisations problématiques
tf.constant(False, dtype=tf.bool)

from tensorflow import keras
import keras_cv
from google.colab import drive
from matplotlib import pyplot as plt
from PIL import Image

# Disable the traceback filtering for more verbose error messages
keras.config.disable_traceback_filtering()

model = keras_cv.models.StableDiffusion(
    img_width=512,
    img_height=512,
    jit_compile=True,
)

images = model.text_to_image(prompt='an electric guitar sculture by Rodin, highly detailed, in clay', batch_size=1)
plt.imshow(images[0])
plt.axis('off')

drive.mount('content/drive')

Image.fromarray(images[0]).save('content/drive/My Drive/google_colab/image.png')

```

```python
import time
import keras_cv
import keras
import matplotlib.pyplot as plt
```

```python
model = keras_cv.models.StableDiffusion(
    img_width=512, img_height=512, jit_compile=False
)
```

```python
images = model.text_to_image(prompt="photograph of an astronaut riding a horse", batch_size=1)


def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")


plot_images(images)
```
